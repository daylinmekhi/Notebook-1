## 3 - Making Functions

Ok, let's look at our chatbot so far. Below we have added three different requests as a nested if statement that the chatbot can handle requesting vacation, contact, and location information.

Read the code below, run it and test it.


````python
print("Hi. What's your name?")
name = input()

if "Stev" in name:
  print("I'm sorry, " + name + ", your system access has been revoked.")
else:
  # Ask user for input
  request = input("Hello, " + name + ". Which would you like help with? 'vacation', 'contact', or 'location'. ")

  # Handle different requests
  if "vacation" in request:
    print("Your allowance is 15 vacation days per year, and they must be booked in the vacation system at least 3 weeks in advance." )
  elif "contact" in request:
    print("You can contact HR at help@pycorp.com")
  elif "location" in request:
    print("The office is located at 404 Coding Ave, Pyville")
  else:
    print("I'm sorry, this chatbot is currently under development and can provide information on vacation, contactn and location.")
````

**Note: the green pieces of text above are not code, they are comments. In Python the `#` symbol tells the interpreter to ignore that line of code, so you can write in any text you want. Comments are *really important* because they explain what your code does, which makes it so much easier to edit in the future.** At the moment our code is kind of simple, so easier to read, but as our code gets more complex, comments will be vital.

Now, what would we need to add to this code if, after we've answered a user's first request, we asked if they'd like to make a second request? With our current knowledge of code, it would look something like the below.

Read through the code to understand it, then run it and test it:

````python
name = input("Hi. What's your name?")

if "Stev" in name:
  print("I'm sorry, " + name + ", your system access has been revoked.")
else:
  # Ask users for input and handle different request
  request = input("Hello, " + name + ". Which would you like help with? 'vacation', 'contact', or 'location'. ")

  if "vacation" in request:
    print("Your allowance is 20 vacation days per year, and they must be booked in the vacation system at least 3 weeks in advance." )
  elif "contact" in request:
    print("You can contact HR at help@pycorp.com")
  elif "location" in request:
    print("The office is located at 404 Coding Ave, Pyville")
  else:
    print("I'm sorry, this chatbot is currently under development and can provide information on vacation, contactn and location.")

  # Ask user if they want still want help
  request = input("Is there anything else I can help you with? (y/n)")
  if "y" in request:

    # Repeat the code from above
    request = input("Do you want help with something else?")

    if "vacation" in request:
      print("Your allowance is 15 vacation days per year, and they must be booked in the vacation system at least 3 weeks in advance." )
    elif "contact" in request:
      print("You can contact HR at help@pycorp.com")
    elif "location" in request:
      print("The office is located at 404 Coding Ave, Pyville")
    else:
      print("I'm sorry, this chatbot is currently under development and can provide information on vacation, contactn and location.")

  else:
    print("Thank you for using this chatbot!")
````

Ok, so how does that help us with our chatbot?

Below, we've started by defining a function called `handle_request()` that asks a user what they want help with, and gives them the corresponding answer.

Then, we've written the code for running our chatbot. Read through the code below, try to understand it, and then experiment with it to understand how it works:

````python
# Define a function that handles all requests
def handle_request():
  request = input("Which would you like help with? 'vacation', 'contact', or 'location': ")

  # Handle different requests
  if "vacation" in request:
    print("Your allowance is 15 vacation days per year, and they must be booked in the vacation system at least 3 weeks in advance." )
  elif "contact" in request:
    print("You can contact HR at help@pycorp.com")
  elif "location" in request:
    print("The office is located at 404 Coding Ave, Pyville")
  else:
    print("I'm sorry, this chatbot is currently under development and can provide information on vacation, contact and location.")


# Check the user
name = input("Hi. What's your name? ")
if "Stev" in name:
    print("I'm sorry, " + name + ", your system access has been revoked.")
else:
    # First request
    print("Access granted. Welcome " + name + "!")
    handle_request()

    # Second request
    request = input("Is there anything else you'd like help with? (y/n) ")
    if "y" in request:
      handle_request()
    else:
      print("Thank you for using this chatbot!")
````

Notice that `handle_request()` is called twice, which means we don't have to rewrite any code.

So functions are useful for **not rewriting** code, but they also make your code more **flexible** - it's easier to be creative, and try different solutions, when you only have to type one word, instead of copy and pasting whole sections of code, and also more **readable**.
