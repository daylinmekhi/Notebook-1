## 1 - Modularization

Fancy word, huh? Well, "Modularization" is a simple, but really important concept: **always package your code into clear, simple functions.**

To see why this is valuable, let's start this lab by looking at the final code for our chatbot from Lab 4:

````python
# Define a function that handles all requests
def handle_request():
  request = input("Which would you like help with? 'vacation', 'contact', or 'location': ")

  # Handle different requests
  if "vacation" in request:
    print("Your allowance is 15 vacation days per year, and they must be booked in the vacation system at least 3 weeks in advance." )
  elif "contact" in request:
    print("You can contact HR at help@pycorp.com")
  elif "location" in request:
    print("The office is located at 404 Coding Ave, Pyville")
  else:
    print("I'm sorry, this chatbot is currently under development and can provide information on vacation, contact and location.")


# Check the user
name = input("Hi. What's your name? ")
if "Stev" in name:
    print("I'm sorry, " + name + ", your system access has been revoked.")
else:
    # First request
    print("Access granted. Welcome " + name + "!")
    handle_request()

    # Second request
    request = input("Is there anything else you'd like help with? (y/n) ")
    if "y" in request:
      handle_request()
    else:
      print("Thank you for using this chatbot!")
````

We defined one function, but there are a couple of other opportunities to use functions that we missed.

````python
def handle_request():
    request = input("Which would you like help with? 'vacation', 'contact', or 'location': ")

    if "vacation" in request:
        print("Your allowance is 15 vacation days per year, and they must be booked in the vacation system at least 3 weeks in advance.")
    elif "contact" in request:
        print("You can contact HR at help@pycorp.com")
    elif "location" in request:
        print("The office is located at 404 Coding Ave, Pyville")
    else:
        print("I'm sorry, this chatbot is currently under development and can provide information on vacation, contact and location.")

def check_user(name):
    if "Stev" in name:
        print("I'm sorry, " + name + ", your system access has been revoked.")
        return False
    else:
        print("Access granted. Welcome " + name + "!")
        return True

def chatbot():
    name = input("Hi. What's your name? ")
    if check_user(name):
        handle_request()
        another_request = input("Is there anything else you'd like help with? (y/n) ")
        if "y" in another_request:
            handle_request()
        else:
            print("Thank you for using this chatbot!")

chatbot()
````

One of the advantages of modularization is that it is easier to edit the functionality of our code.

**Problem:** we want to repeat code until a certain condition is met

**Soution:** `while` loops

The basic format of a `while` loop is written below. Read through the code to see if you can understand how it works, and then run it and see if you can make it exit it's loop.

````python
while True:
  request = input("Which would you like help with? 'vacation', 'contact', 'location', or 'exit': ")

  if request == "vacation":
      print("Your allowance is 15 vacation days per year, and they must be booked in the vacation system at least 3 weeks in advance.")
  elif request == "contact":
      print("You can contact HR at help@pycorp.com")
  elif request == "location":
      print("The office is located at 404 Coding Ave, Pyville")
  elif request == "exit":
      print("Thank you for using this chatbot!")
      break
  else:
      print("I'm sorry, this chatbot is currently under development and can provide information on vacation, contact, and location.")
````

In short, the code "inside" the loop keeps running from top to bottom, until the condition stated after "while" is False. In Python, the quickest way to write that inside a function is to use `break`.

**Solution:** Dictionaries

We have already encountered different types of data in the course, for example strings (text) and numbers. Dictionaries are another type of data, but it is one that stores smaller pieces of data.

**Solution:** using file reading functions.

Below, we'll explain how you can use code to access information from a `.csv`

What we mean by that is that reading data from `.csv` files is not built-in to Python, like loops and conditionals are. We can still do it, but we need to **import a package of functions to do it**.

Packages are something we'll cover in more detail in future weeks, but for now all you need to know is:

In the code cell below, we've written some code that reads the `policies.csv` in the shared folder with this file, and produces a dictionary. Run it to check that it works!

Note: the name of the package we are using is a really popular one called `pandas`. To learn more about it and the creator of it, [read this story](https://qz.com/1126615/the-story-of-the-most-important-tool-in-data-science). If you want to understand how we knew to do that, you can just google "load csv file in python"!



````python
import pandas

def load_responses_from_csv(filename):
    df = pandas.read_csv(filename)
    responses = df.set_index('keyword')['response'].to_dict()
    return responses

responses = load_responses_from_csv('policies.csv')
print(responses)
````

There's a few components to explain here, so below is the same code, but with comments explaining what each line does:

````python
# The next line imports the package, to be used later
import pandas

def load_responses_from_csv(filename):
    # Read the CSV file into it's raw format (called a DataFrame)
    df = pandas.read_csv(filename)
    # Convert the DataFrame into a dictionary
    responses = df.set_index('keyword')['response'].to_dict()
    # output the dictionary
    return responses

responses = load_responses_from_csv('policies.csv')
print(responses)
````

The code below writes a new policy to the csv file about paternity leave. run it, and then re-run your chatbot above to see if it worked!

````python
import pandas

filename = "policies.csv"
keyword = "paternity"
response = "Paternity leave is 4 months, and may be taken at any time within the first 12 months after the birth of your child."

def write_new_policy_to_csv(filename, keyword, response):
    new_policy = {'keyword': [keyword], 'response': [response]}
    df_new_policy = pandas.DataFrame(new_policy)
    df_new_policy.to_csv(filename, mode='a', header=False, index=False)

write_new_policy_to_csv(filename, keyword, response)
````
